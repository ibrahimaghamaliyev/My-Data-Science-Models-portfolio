```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = data_high_corr[[#'Week of the month',
                            'Day of the week',
                            #'Non-urgent order',
                            #'Urgent order',
                            #'Order type A',
                            #'Order type B',
                            #'Order type C',
                            #'Orders from the traffic controller sector',
                            'Banking orders (1)',
                            'Banking orders (2)',
                            'Banking orders (3)'
                            ]]

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns

vif = vif.sort_values(by = ['VIF'], ascending = False)
vif
```

```python
data_high_vif = data_high_corr[['Banking orders (3)', 'Day of the week', 'Banking orders (2)', 'Banking orders (1)', 'Target (Total orders)']]
```

```python
for i in data_high_vif.columns:
  sns.boxplot(data = data_high_vif, x = data_high_vif[i])
  plt.show()
```

```python
q1 = data_high_vif.quantile(0.25)
q3 = data_high_vif.quantile(0.75)
IQR = q3 - q1

Upper = q3 + 1.5*IQR
Lower = q1 - 1.5*IQR
```

```python
for i in data_high_vif.columns:
  data_high_vif[i] = np.where(data_high_vif[i] > Upper[i], Upper[i], data_high_vif[i])
  data_high_vif[i] = np.where(data_high_vif[i] < Lower[i], Lower[i], data_high_vif[i])
  sns.boxplot(data = data_high_vif, x = data_high_vif[i])
  plt.show()
```

```python
inputs = data_high_vif.drop(['Target (Total orders)'], axis = 1)
output = data_high_vif['Target (Total orders)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.2, random_state = 42)
```

```python
from sklearn import metrics
```

```python
def evaluate(model, X_test, y_test):

    y_pred = model.predict(X_test)

    mae_test= metrics.mean_absolute_error(y_test,y_pred)
    mse_test=metrics.mean_squared_error(y_test,y_pred)
    rmse_test= np.sqrt(mse_test)
    r2_test= metrics.r2_score(y_test,y_pred)


    y_pred_t= model.predict(X_train)

    mae_train=metrics.mean_absolute_error(y_train,y_pred_t)
    mse_train=metrics.mean_squared_error(y_train,y_pred_t)
    rmse_train=np.sqrt(mse_train)
    r2_train= metrics.r2_score(y_train,y_pred_t)


    result={
        'metric':['mae','mse','rmse','r2'],
        'train':[mae_train,mse_train,rmse_train,r2_train*100],
        'test':[mae_test,mse_test,rmse_test,r2_test*100]
    }


    results= pd.DataFrame(result)

    print(results)
```

```python
from sklearn.svm import SVR
first_model = SVR()
first_model.fit(X_train, y_train)
first_result = evaluate(first_model, X_test, y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV

kernel = ['linear', 'poly', 'rbf', 'sigmoid']
gamma = ['scale', 'auto']
C = [1, 10, 100, 1e3, 1e4, 1e5, 1e6]
epsilon = [0.1, 0.01, 0.001, 0.0001]

random_grid = {'kernel' : kernel,
               'gamma' : gamma,
               'C' : C,
               'epsilon' :epsilon}

print(random_grid)
```

```python
svr_random = RandomizedSearchCV(estimator = first_model, param_distributions = random_grid, n_iter = 1, cv = 3, verbose = 1, n_jobs = -1)
svr_random.fit(X_train, y_train)
```

```python
svr_random.best_params_
```

```python
optimized_model = svr_random.best_estimator_
optimized_accuracy = evaluate(optimized_model, X_test, y_test)
```

```python
variables = []
train_r2_scores = []
test_r2_scores = []

for i in X_train.columns:
    X_train_single_var = X_train[[i]]
    X_test_single_var = X_test[[i]]

    optimized_model.fit(X_train_single_var, y_train)
    y_pred_train_single_var = optimized_model.predict(X_train_single_var)

    train_r2 = metrics.r2_score(y_train, y_pred_train_single_var)

    y_pred_test_single_var = optimized_model.predict(X_test_single_var)

    test_r2 = metrics.r2_score(y_test, y_pred_test_single_var)

    variables.append(i)
    train_r2_scores.append(train_r2)
    test_r2_scores.append(test_r2)


results_df = pd.DataFrame({'Variable': variables, 'Train R2': train_r2_scores, 'Test R2': test_r2_scores})

results_df_sorted = results_df.sort_values(by='Test R2', ascending=False)

pd.options.display.float_format = '{:.2f}'.format

results_df_sorted
```
